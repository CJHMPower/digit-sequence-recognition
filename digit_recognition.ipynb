Randomly selected a series of digit and concatate them into one picture.


```python
import numpy as np

n_class, n_len, width, height = 11, 5, 28, 28          #n_class=11 because there may be a blank

def generate_dataset(X, y):
    X_len = X.shape[0]
    X_gen = np.zeros((X_len, height, width*n_len, 1), dtype=np.uint8)
    y_gen = [np.zeros((X_len, n_class), dtype=np.uint8) for i in range(n_len)]
    # generate random numbers of digits
    n_digit=random.randint(1,6)
    for j in range(X_len):
        n_digit=random.randint(1,5)
        for i in range(n_digit):
            index=random.randint(0,X_len-1)
            image=X[index]
            label=y[index]
            X_gen[j][:,i*width:width+i*width,0]=image
            y_gen[i][j][label]=1
        for i in range(n_digit,n_len):
            y_gen[i][j][10]=1
    return X_gen, y_gen

X_train, y_train = generate_dataset(X_raw, y_raw)
X_test, y_test = generate_dataset(X_raw_test, y_raw_test)
X_raw=X_raw.reshape(n_train,width,height,1)
Y_raw=np.zeros((n_train,n_class))
for i in range(n_train):
    Y_raw[i][y_raw[i]]=1
```

```python
# visualize the digit sequence generated
for i in range(15):
    plt.subplot(5, 3, i+1)
    index = random.randint(0, n_test-1)
    title = ''
    for j in range(n_len):
        title += str(np.argmax(y_test[j][index])) + ','
    
    plt.title(title)
    plt.imshow(X_test[index][:,:,0], cmap='gray')
    plt.axis('off')
    
```

Here is the model design.
I used convolution neural network, pool layer, fully connected output layer and Dropout technology.
Convolution neural network is a feedforward neural network, which can directly input the original image as an input, with a convolution kernel to extract the local pixels in the image, by scanning the entire image can be mapped to different depth and length and width of the size of the other An image, and by weight sharing can greatly reduce the number of parameters. 
After the convolution network is obtained, MaxPooling is used to extract the neurons with the largest local weights. This is to prevent over-fitting, and in the middle of the process, the Dropout technique is used to discard some weights with a certain probability. Fit and make the neural network more robust.

```python
from keras.models import Model
from keras.optimizers import SGD
from keras.models import Sequential
from keras.layers import *
a=Input(shape=(height,n_len*width,1))

d1=16        #depth of filter kernel each layer
d2=16
d3=64
d4=128
d5=256


drop_out=0.25
patch_size=3             
k_size=(2,2)

conv1=Convolution2D(nb_filter=d1,nb_row=patch_size,nb_col=patch_size,border_mode='same',activation='relu')(a)
conv1=MaxPooling2D(pool_size=k_size,border_mode='same')(conv1)


conv2=Convolution2D(nb_filter=d2,nb_row=patch_size,nb_col=patch_size,border_mode='same',activation='relu')(conv1)
conv2=MaxPooling2D(pool_size=k_size,border_mode='same')(conv2)


conv3=Convolution2D(nb_filter=d3,nb_row=patch_size,nb_col=patch_size,border_mode='same',activation='relu')(conv2)
conv3=MaxPooling2D(pool_size=k_size,border_mode='same')(conv3)


conv4=Convolution2D(nb_filter=d4,nb_row=patch_size,nb_col=patch_size,border_mode='same',activation='relu')(conv3)
conv4=MaxPooling2D(pool_size=k_size,border_mode='same')(conv4)



conv5=Convolution2D(nb_filter=d5,nb_row=patch_size,nb_col=patch_size,border_mode='same',activation='relu')(conv4)
conv5=MaxPooling2D(pool_size=k_size,border_mode='same')(conv5)



x=Flatten()(conv5)


x=Dropout(drop_out)(x)
node=32

x_1=Dense(node,activation='relu')(x)          #connect the flatten layer to five classifier,each one comes to a digit.
x_2=Dense(node,activation='relu')(x)
x_3=Dense(node,activation='relu')(x)
x_4=Dense(node,activation='relu')(x)
x_5=Dense(node,activation='relu')(x)


d1=Dense(n_class,activation='softmax')(x_1)
d2=Dense(n_class,activation='softmax')(x_2)
d3=Dense(n_class,activation='softmax')(x_3)
d4=Dense(n_class,activation='softmax')(x_4)
d5=Dense(n_class,activation='softmax')(x_5)

outputs=[d1,d2,d3,d4,d5]


model=Model(input=a,output=outputs)
model.compile(loss='categorical_crossentropy', optimizer='adadelta',metrics=['accuracy'])  
    
#final output layer,to minimize the categorial crossentropy
          


```

### 练习：训练你的网络模型

你需要设置20%的数据作为验证集，以保证模型没有过拟合。

```python
#Train the neural network.This cell may consume 20 minutes to run on an i7-6700HQ CPU.

model.fit(X_train,y_train,batch_size=200,nb_epoch=5,validation_split=0.2)
```

```python
#Compute the total accuracy.An image is seen as correctly classified only when all digits in it are classified correctly.

def evaluate(model):
    global y_test,X_test
    y_true=np.array(y_test)
    y_true=np.argmax(y_true,axis=2).T
    pred=np.array(model.predict(X_test))
    pred=np.argmax(pred,axis=2).T
    acc=np.mean(np.sum(np.equal(pred,y_true),axis=1)==5)
    return acc
        
    
    

print(evaluate(model))
```

We can see that the final accuracy for all images on testing set is 93.45,useful and applicable for the real application.

Visualize the real prediction results.

```python

def get_result(result):
    resultstr = ''
    for i in range(n_len):
        resultstr += str(np.argmax(result[i])) + ','
    return resultstr

index = random.randint(0, n_test-1)
y_pred = model.predict(X_test[index].reshape(1, height, width*n_len, 1))

plt.title('real: %s\npred:%s'%(get_result([y_test[x][index] for x in range(n_len)]), get_result(y_pred)))
plt.imshow(X_test[index,:,:,0], cmap='gray')
plt.axis('off')
```
